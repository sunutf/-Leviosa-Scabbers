```
import numpy as np
from matplotlib import pyplot
import pdb
from scipy import stats


total_samples = 1000 
num_samples = 12
approximate_gaussian = []
gaussian = []
for i in range(total_samples):
	sample_ = np.random.uniform(0,1,num_samples)
	sample_ = np.sum(sample_) - 6.0
	approximate_gaussian.append(sample_)
	gaussian.append(np.random.randn())
bins = np.linspace(-5,5,100)
pyplot.hist(approximate_gaussian, bins,alpha=0.5, label="Uniform samples",color='r')
pyplot.hist(gaussian, bins, alpha=0.5, label="Gaussian samples", color='b')
pyplot.legend(loc='upper left')
pyplot.show()
```

```
## 1. 실제 정규분포
X = np.arange(-5,5,0.1)
Y = stats.norm.pdf(X, loc=0, scale=1.0)

## 2. 샘플 세트 2개 추출
total_samples = 1000
num_samples = 12
sample_set = [[],[]]
for i in range(total_samples):
  for j in range(2):
    sample_ = np.random.uniform(0,1,num_samples)
    sample_ = np.sum(sample_) - 6.0
    sample_set[j].append(sample_)
  

## 3. 샘플 세트 2개에 대해 가우시안 커널 밀도 추정
estimator0 = stats.gaussian_kde(sample_set[0], bw_method='silverman')
estimator1 = stats.gaussian_kde(sample_set[1], bw_method='silverman')
K_0 = estimator0(X)
K_1 = estimator1(X)

## 4. 그래프 그리기
bins = np.linspace(-5,5,100)

pyplot.hist(sample_set[0], bins, alpha=0.5, label="Uniform sample 0", color='r', density=True)
pyplot.hist(sample_set[1], bins, alpha=0.5, label="Uniform sample 1", color='b', density=True)
pyplot.plot(X, Y, label='Original', color='black')
pyplot.plot(X, K_0, label='kde_sample_0')
pyplot.plot(X, K_1, label='kde_sample_1')

pyplot.legend(loc='best')
pyplot.show()
```
